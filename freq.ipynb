```python
import csv

# 'db/nr30_chains/mod_CG/feat_4_freq_NRaa_mod2_8.csv'
# 'feat_4_freq_NRaa_mod2_8_nor1_test.csv'

def get_freq(file,n):
    with open(file,'r') as f:
        r = csv.reader(f)
        next(r)
        freq = [0]*n
        for line in r:
            freq[int(line[0])] += 1
    return freq
    
```

```python
import matplotlib.pyplot as plt

# 'db/nr30_chains/mod_CG/feat_4_freq_NRaa_mod2_8_nor1.csv'
csv_file = 'db/nr30_chains/feat_4_freq_NRaa_nor1.csv'
num_cg = 16

plt.figure(figsize=(8,5))
x_ax = list(range(2,num_cg+1))


f1 = get_freq(csv_file, num_cg-1)  # num_cg-1 since this is no-r1 database
norm_f1 = [i/sum(f1) for i in f1]
plt.plot(x_ax, norm_f1)

# f2 = get_freq('feat_4_freq_NRaa_mod2_8_nor1_test.csv',18)
# norm_f2 = [i/sum(f2) for i in f2]
# plt.plot(x_ax, norm_f2)


plt.xticks(x_ax)
plt.ylabel("CG frequency (%)")
plt.xlabel("CG number")
plt.title("CG distribution in NRaa (4Ã…) - No r1 database")
plt.show()
```
